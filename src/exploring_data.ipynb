```python
data_dir = os.path.join("drive", "MyDrive", "cars")
train_dir = os.path.join(data_dir, "train")
val_dir = os.path.join(data_dir, "val")
```

```python
classes = sorted(os.listdir(train_dir))
n_classes = len(classes)
assert classes == sorted(os.listdir(val_dir))
```

```python
print("Total number of classes: ", n_classes)
for i, class_ in enumerate(classes):
  print("#", i+1, "\t", class_)
```

```python
colors = pd.Series(classes)

n_train, n_val = list(), list()
for class_ in classes:
  n_train.append(len(os.listdir(os.path.join(train_dir, class_))))
  n_val.append(len(os.listdir(os.path.join(val_dir, class_))))

n_train = pd.Series(n_train)
n_val = pd.Series(n_val)
```

```python
df_dict = {
           "color": colors,
           "train": n_train,
           "validation": n_val,
           "total": n_train + n_val,
           "trainx0.2": n_train * .2,
           "combined-in-train": n_train * .2 <= n_val
           }
df = pd.DataFrame(df_dict)
df
```

Above colors include modified images from both datasets (i.e. kaggle 
[VCoR](https://www.kaggle.com/datasets/landrykezebou/vcor-vehicle-color-recognition-dataset) and [Shenasa](https://shenasa-ai.ir/))

And other colors just include modeified images from kaggle VCoR dataset. Becuase I want the validation set, only includes Shenas images.
